# Going Beyond Groups and into the World of Individuals: Estimating Experimental Effects on Individuals by the Changes of their Data Distributions

## Abstract


Traditionally, experimental effects on humans are investigated at the group level. However, taking individuality into account is sometimes essential, such as when investigating the behavior of a diverse group or when diagnosing and treating individuals for psychiatric disorders. In this work, we present a distributional "Spotlight" for individual-level investigations that measures the effects on individuals through the changes of the probability distributions of their data. As our testbed, we examine two well-known effects on human processing speed -- (1) the effect of information set sizes and (2) the effect of irrelevant information. The effects are measured by the changes in reaction time (RT) distributions using three methods. Our Spotlight approach, in which we directly measure the changes of the information contained in the distributions, is validated by comparing against the distributional changes captured by two traditional models of RT: the ex-Gaussian distribution and the Drift-Diffusion model. Our results reveal large individual differences of the effects for both set size and irrelevant information. Some differences can be explained by the individual differences of information processing strategies (such as sequential processing vs chunked processing) and in information selection. Generally, our results highlight the need to investigate experimental effects at the individual level and the promise of our distributional Spotlight in such pursuits. We discuss applications in two research areas that may greatly benefit from individual-level investigations: human information processing and precision psychiatry.

**Keywords**: Information Processing; Precision Psychiatry; Individual Differences; Reaction Time Distributions; Drift-Diffusion Model; Choice Reaction; Visual Search; Set Size Effect; Simon Effect

## When to go beyond the mean?

Many researchers across fields in different eras urged to go beyond the mean. But adoption remains slow, perhaps because the reasons to switch are not well understood/known. We try to draw a clear line as to when and not it is appropriate to use the mean for any type of data. 

When we say the ``mean'' or the average, we refer specifically to the mean of a normal distribution. So, the underlying assumption we make while averaging is that our data is normally distributed. A lot of naturally occurring data follows the normal distribution (from human heights, weights, IQ scores, to material properties and stock prices and down to brownian motion of particles). The mean is specifically the expected value of a normal distribution. Its significance is well-studied, let us still illustrate with a simple example. If we are to bet on a value that is normally distributed and we get to play many times, our best bet is to keep betting on the mean and in the long run, we are guaranteed to maximize our profits relative to what we know (i.e., we only know that the data is normally distributed and no other information). Being the expected value, the mean is also our best prediction in such cases.

Evidently, the mean contains important information about normally distributed data. However, it is the central limit theorem that truly makes the use of normal distribution integral to experimental research. The theorem states that if we repeatedly sample from a population, our sample means would follow a normal distribution. Therefore, when we collect many samples from the population for our research, we can expect the data to be normally distributed.

But what about the data distributions from each individual? The data distributions may considerably differ from normal distributions. One prime example is the RT distributions of individuals in information processing tasks. RT distributions have been generally found to be asymmetric or skewed, unlike normal distributions which are symmetric. To account for the skewness, researchers have adopted different distributions -- notably, gamma distributions, exponentially modified normal distributions, and Weibull distributions. Importantly, we do not only risk missing out on the skewness information if we model RT data with normal distributions. Rather, for non-normal distributions the mean is not the expected value and the whole distribution would not represent the actual data well. Therefore, it is extremely important to consider the actual distribution of the data.

Before diving into distributional analyses, a relevant question is: Is non-normality a
rarity or the norm? Several researchers over the last four decades have asked this question for psychological experimental data. In a meta-analysis of 693 distributions, Blanca et al. (2013) found that 95% of the data violated the assumptions of normality. Previously, Micceri (1989) tested 440 datasets for normality and found all of them to be significantly non-normal. Put simply, evidence shows that very rarely our experimental data is close to being normal, so much so that Micceri (1989) dubbed the normal distribution as the "unicorn" in psychological research.

## Distributional Changes as Experimental Effects on Individuals

What should we then if we cannot use the normal distributions? A general approach that has been promising is investigating whole distributions of data and quantify their changes as the experimental effect. However, there are some challenges to this approach despite its success. Typically the distributions are explained by multi-parameter models, which makes investigated the changes complicated and quite complex when want a model for each individual. For example, one approach is modeling the RT data using a parametric distribution (e.g., Gamma, Exponential Normal) and investigate the distributions based on the parameters. In some cases, we may have a theoretical model that produces the distributions when simulated. In such cases, we use the model parameters to investigate the experimental effects. The Drift-Diffusion Model (DDM) for binary-choice tasks provide an example.

In this work, we explore a simple and general way to examine the experimental effects on individuals by the changes of the distributions. Specifically, we measure the distributional changes directly by the changes of information contained in the RT distributions. Our goal is to highlight and magnify the changes -- therefore, we refer to this method as our "Spotlight" method. We compare our Spotlight against two more traditional methods -- (1) modeling RT data with exponentially modified Gaussian (Ex-G) distributions, and (2) modeling RT data with the DDM. Please see Balota et al (2008) for a discussion on using the Ex-G distributions for RT. Elaborate discussions about the DDM can be found in many articles over the years; please see Voss et al (2004, 2013) and Shinn (2020) for relatively recent reviews.

We test our distributional Spotlight on two well-known effects on human information processing: (1) the effect of information set size and (2) the effect of irrelevant information in the set. Typically, the average subject demonstrates both of these effects -- that is, RT is usually slower with increasing set size and increasing amount of irrelevant information. These results indicate that that generally (1) humans are serial processors of information and (2) humans are unable to inhibit or ignore irrelevant information.

However, do all individuals experience the same effects? We examine this question in two separate studies using the three distributional methods described above.

# Study 2: Setsize effect vs Simon Effect in Choice Reaction Time Tasks

## Experimental Task and RT Dataset:

We use the RT dataset from Experiment 2 by Mittelstadt and Miller (2020). This experiment used a 2x2 within-subjects design in which 65 individuals performed the RT task under two choice conditions (2-choice vs 4-choice) and two S-R compatibility conditions (Congruent vs Incongruent). In each trial, the participant is shown a stimulus shown on a computer screen and their task is to make the right response as soon as possible based on the color of the stimulus shown using keyboard. 

The stimulus appeared in one of two locations, left or right from the center of the screen. In 2-choice tasks, a participant would respond using 2 keys ("Z" and "M", placed left and right on a QWERTY keyboard). Each key corresponded to one of two possible colors (e.g., "Z" for Red stimulus and "M" for Blue stimulus). In 4-choice tasks, a participant would respond using 4 keys ("Z", "X", "N", and "M"; the first two placed left and the second two placed right on a QWERTY keyboard). Each key corresponded to one of four possible colors. A trial is compatible if the stimulus location is on the same side as the correct response key and incompatible if the location is on the opposing side of the correct response key.

Each participant completed 12 experimental blocks in total, six blocks per set size and 64 trials per block. In each block, set size was held constant, but the S-R mappings were varied randomly in equal proportion between compatible and incompatible across trials. Therefore, one block of 64 trials contained 32 trials in compatible mapping condition and 32 incompatible ones presented in a random order completed.

The original dataset can be found in: https://osf.io/nhy4hu.

Summary of dataset:
    
    1. 65 individual participants + 1 "average" subject (by taking the average performance of the 65 individuals)

    2. For each individual, we have a total of 12*64 = 768 samples of RT and 384 samples for each of the 4 conditions in the 2x2 design (2 choices vs 4 choices and compatible vs incompatible S-R mappings)

## ---- Study 2 ----

Here, we -

    Load all data for all participants
    Then, combine all different files into one aggregated dataset ("agg_df")
    In the DataFrame, each row contains all information we have on one individual. 
    
    For each individual, we calculate: 
        (1) Ex-Gaussian fit parameters
        (2) RE between the Ex-Gaussian distributions across conditions 
              --> an RE matrix comparing the distributions of all conditions with each other. 
        (3) Drift-diffusion model parameters by fitting RT distributions from each condition.
        
        (4) Mean Accuracy across conditions.

```python
# Hyper-Parameters

new_run = 0 # NOTE: make 1 to rerun from scratch
num_subj_exp2 = 65
list_of_actual_subj_num_exp2 = list(np.arange(1,29)) + [30] + list(np.arange(48,84)) # To be discarded after loading the datafiles.
	# Need it as the IDs are not sequential.
	# To be discarded after loading the datafiles.


# Make the first version of agg_data

if (new_run):
    data_exp2_subj = [0]*num_subj_exp2
    
    for i_subj_exp2, i_subj_exp2_actual in zip(range(num_subj_exp2), list_of_actual_subj_num_exp2):
        data_exp2_subj[i_subj_exp2 - 1] = pd.read_csv("Datasets/OurStudy2_Mittelstadt_exp2/RawData_Experiment 2/RawData_Experiment 2/M1509-{0:03}.Txt".format(i_subj_exp2_actual), skiprows = 3, delim_whitespace=True)
        data_exp2_subj[i_subj_exp2 - 1].columns = ["i_sub", "i_block", "i_trial", "trty", "cor", "rt"]

    # n2 and n4: 2-choice vs 4-choice
    # c1 and c2: congruent vs incongruent
    agg_data = pd.DataFrame(columns = ["pid", "rt_n2c1", "rt_n2c2", "rt_n4c1", "rt_n4c2", \
                                        "cor_n2c1", "cor_n2c2", "cor_n4c1", "cor_n4c2", \
                                        "sorted_cor_n2c1", "sorted_cor_n2c2", "sorted_cor_n4c1", "sorted_cor_n4c2", \
                                        "trty_n2c1", "trty_n2c2", "trty_n4c1", "trty_n4c2", \
                                        "ddm_params_n2c1", "ddm_params_n2c2", "ddm_params_n4c1", "ddm_params_n4c2"])
                                        # "e_n2c1", "e_n2c2", "e_n4c1", "e_n4c2", \
                                        # "re_n4c1_wrt_n2c1", "re_n4c2_n2c2", \
                                        # "exg_params_n2c1", "exg_params_n2c2", "exg_params_n4c1", "exg_params_n4c2"])

    agg_data.loc[:, "pid"] = np.arange(1, num_subj_exp2 +1) # We are using our own ids. 
                                             # Why? Original ids too complicated, but easy to match as the participant order remains the same for the two sets of ids.
    agg_data.set_index("pid", inplace = True)

     # ---------------- Define Drift Diffusion Model here -----------------

     # Here we define our drift diffusion model.
     # we use four parameters: drift, diffusion or noise, boundary separation, and non-decision time (i.e., time for perception and response execution).
     # We will fit the model to each participant's data to get the parameter estimates for the participant.
     # Note that the time unit used in the pyddm package is SECONDS.
     
    model_ddm = Model(name='Exp2_data',
        drift=DriftConstant(drift = Fittable(minval=0, maxval=20)),
        noise=NoiseConstant(noise=Fittable(minval=.1, maxval = 1.5)),
        bound=BoundConstant(B = Fittable(minval=.1, maxval = 1.5)),
        overlay=OverlayChain(overlays=[OverlayNonDecision(nondectime=Fittable(minval=0, maxval=.4)),
                                        OverlayPoissonMixture(pmixturecoef=.02,
                                                                rate=1)]),
        dx=.001, dt=.001, T_dur=2)


    # ------------------------------------------------------------------------------------------
    # Completing agg data with rt, cor (correct or not), sorted cor (by rt) and try (trial type)

    for i_subj_exp2 in range(1, num_subj_exp2 + 1):
        y = data_exp2_subj[i_subj_exp2-1]

        y_n2c1 = y.loc[((y.i_block == 2) | (y.i_block == 3) | (y.i_block == 4) | (y.i_block == 5) | (y.i_block == 6) | (y.i_block == 7)) & \
                        ((y.trty == 1) | (y.trty == 2) | (y.trty == 5) | (y.trty == 6) )]
        agg_data.at[i_subj_exp2, "rt_n2c1"] = np.array(y_n2c1.rt)
        agg_data.at[i_subj_exp2, "cor_n2c1"] = np.array(y_n2c1.cor)
        agg_data.at[i_subj_exp2, "sorted_cor_n2c1"] = np.array(y_n2c1.loc[:,["rt", "cor"]].sort_values(by = ["rt"]).cor)
        agg_data.at[i_subj_exp2, "trty_n2c1"] = np.array(y_n2c1.trty)

        y_n2c2 = y.loc[((y.i_block == 2) | (y.i_block == 3) | (y.i_block == 4) | (y.i_block == 5) | (y.i_block == 6) | (y.i_block == 7)) & \
                        ((y.trty == 3) | (y.trty == 4) | (y.trty == 7) | (y.trty == 8) )]
        agg_data.at[i_subj_exp2, "rt_n2c2"] = np.array(y_n2c2.rt)
        agg_data.at[i_subj_exp2, "cor_n2c2"] = np.array(y_n2c2.cor)
        agg_data.at[i_subj_exp2, "sorted_cor_n2c2"] = np.array(y_n2c2.loc[:,["rt", "cor"]].sort_values(by = ["rt"]).cor)
        agg_data.at[i_subj_exp2, "trty_n2c2"] = np.array(y_n2c2.trty)

        y_n4c1 = y.loc[((y.i_block == 9) | (y.i_block == 10) | (y.i_block == 11) | (y.i_block == 12) | (y.i_block == 13) | (y.i_block == 14)) & \
                        ((y.trty == 1) | (y.trty == 2) | (y.trty == 5) | (y.trty == 6) )]
        agg_data.at[i_subj_exp2, "rt_n4c1"] = np.array(y_n4c1.rt)
        agg_data.at[i_subj_exp2, "cor_n4c1"] = np.array(y_n4c1.cor)
        agg_data.at[i_subj_exp2, "sorted_cor_n4c1"] = np.array(y_n4c1.loc[:,["rt", "cor"]].sort_values(by = ["rt"]).cor)
        agg_data.at[i_subj_exp2, "trty_n4c1"] = np.array(y_n4c1.trty)

        y_n4c2 = y.loc[((y.i_block == 9) | (y.i_block == 10) | (y.i_block == 11) | (y.i_block == 12) | (y.i_block == 13) | (y.i_block == 14)) & \
                        ((y.trty == 3) | (y.trty == 4) | (y.trty == 7) | (y.trty == 8) )]
        agg_data.at[i_subj_exp2, "rt_n4c2"] = np.array(y_n4c2.rt)
        agg_data.at[i_subj_exp2, "cor_n4c2"] = np.array(y_n4c2.cor)
        agg_data.at[i_subj_exp2, "sorted_cor_n4c2"] = np.array(y_n4c2.loc[:,["rt", "cor"]].sort_values(by = ["rt"]).cor)
        agg_data.at[i_subj_exp2, "trty_n4c2"] = np.array(y_n4c2.trty)


# ------------------------------------------------------------------------------
# ----------------- Drift Diffusion model fitting ------------------------------
# ------------------------------------------------------------------------------

        # ddm_params_list = ["drift", "noise", "bound", "nondectime"] 

        time_start = time.time()

        y_n2c1["rt"] = y_n2c1.rt/1000 # because the unit used in the pyddm package is in SECONDS.
        n2c1_sample_object = Sample.from_pandas_dataframe(y_n2c1, rt_column_name="rt", correct_column_name="cor")
        model_n2c1 = fit_adjust_model(sample= n2c1_sample_object, model=model_ddm, verbose=False)
        params_raw = model_n2c1.parameters()
        n2c1_params = [params_raw["drift"]["drift"] + 0, params_raw["noise"]["noise"] + 0, \
                            params_raw["bound"]["B"] + 0, params_raw["overlay"]["nondectime"] + 0]
        print("n2c1 params --------> ", n2c1_params)
        agg_data.at[i_subj_exp2, "ddm_params_n2c1"] = n2c1_params
        # print(y_n2c1.head())
        # exit()

        y_n2c2["rt"] = y_n2c2.rt/1000 # because the unit used in the pyddm package is in SECONDS.
        n2c2_sample_object = Sample.from_pandas_dataframe(y_n2c2, rt_column_name="rt", correct_column_name="cor")
        model_n2c2 = fit_adjust_model(sample= n2c2_sample_object, model=model_ddm, verbose=False)
        params_raw = model_n2c2.parameters()
        n2c2_params = [params_raw["drift"]["drift"] + 0, params_raw["noise"]["noise"] + 0, \
                            params_raw["bound"]["B"] + 0, params_raw["overlay"]["nondectime"] + 0]
        print("n2c2 params --------> ", n2c2_params)
        agg_data.at[i_subj_exp2, "ddm_params_n2c2"] = n2c2_params

        y_n4c1["rt"] = y_n4c1.rt/1000 # because the unit used in the pyddm package is in SECONDS.
        n4c1_sample_object = Sample.from_pandas_dataframe(y_n4c1, rt_column_name="rt", correct_column_name="cor")
        model_n4c1 = fit_adjust_model(sample= n4c1_sample_object, model=model_ddm, verbose=False)
        params_raw = model_n4c1.parameters()
        n4c1_params = [params_raw["drift"]["drift"] + 0, params_raw["noise"]["noise"] + 0, \
                            params_raw["bound"]["B"] + 0, params_raw["overlay"]["nondectime"] + 0]
        print("n4c1 params --------> ", n4c1_params)
        agg_data.at[i_subj_exp2, "ddm_params_n4c1"] = n4c1_params

        y_n4c2["rt"] = y_n4c2.rt/1000 # because the unit used in the pyddm package is in SECONDS.
        n4c2_sample_object = Sample.from_pandas_dataframe(y_n4c2, rt_column_name="rt", correct_column_name="cor")
        model_n4c2 = fit_adjust_model(sample= n4c2_sample_object, model=model_ddm, verbose=False)
        params_raw = model_n4c2.parameters()
        n4c2_params = [params_raw["drift"]["drift"] + 0, params_raw["noise"]["noise"] + 0, \
                            params_raw["bound"]["B"] + 0, params_raw["overlay"]["nondectime"] + 0]
        print("n4c2 params --------> ", n4c2_params)
        agg_data.at[i_subj_exp2, "ddm_params_n4c2"] = n4c2_params

        print("time taken for subject %i:"%i_subj_exp2, time.time() - time_start)

    # --------------------------------------------------------------------------------
    # -------------------- Work directly on the agg_data file. -----------------------

    # agg_data = agg_data_exp2

    ## NOTE on Parameter Estimation
    # all fits done with MLE. BPE tests were done too. 
    # But was getting weird results for some participants when tried to use the same prior for all.
    # So decided to leave the battle for another day.

    # Return order for Ex-G parameters: k (NOT tau directly), mu, sigma
    agg_data["exg_params_n2c1"] = agg_data.rt_n2c1.apply(stats.exponnorm.fit)
    agg_data["exg_params_n2c2"] = agg_data.rt_n2c2.apply(stats.exponnorm.fit)
    agg_data["exg_params_n4c1"] = agg_data.rt_n4c1.apply(stats.exponnorm.fit)
    agg_data["exg_params_n4c2"] = agg_data.rt_n4c2.apply(stats.exponnorm.fit)

    ## Gamma test. Not bad, at least doesn't seem worse than ex_Gaussian anyway.   
    # agg_data["gam_params_n2c1"] = agg_data.rt_n2c1.apply(stats.gamma.fit)
    # agg_data["gam_params_n2c2"] = agg_data.rt_n2c2.apply(stats.gamma.fit)
    # agg_data["gam_params_n4c1"] = agg_data.rt_n4c1.apply(stats.gamma.fit)
    # agg_data["gam_params_n4c2"] = agg_data.rt_n4c2.apply(stats.gamma.fit)

    # ---------------------  Moments -------------------------------------------------
    # Calculate all moments. Possibly wont used them, but still. However, if we do use it, we need to write about how scipy calculated them, even if in just a sentence.  
    # These can be compared against moments from ex-G or gamma distribution parameters.

    # mean, var, skew, kurt = gamma.stats(a, moments='mvsk') # Moments for Gamma distributions

    agg_data["moments_n2c1"] = agg_data.rt_n2c1.apply(lambda x: [np.mean(x), np.std(x), stats.skew(x), stats.kurtosis(x)])
    agg_data["moments_n2c2"] = agg_data.rt_n2c2.apply(lambda x: [np.mean(x), np.std(x), stats.skew(x), stats.kurtosis(x)])
    agg_data["moments_n4c1"] = agg_data.rt_n4c1.apply(lambda x: [np.mean(x), np.std(x), stats.skew(x), stats.kurtosis(x)])
    agg_data["moments_n4c2"] = agg_data.rt_n4c2.apply(lambda x: [np.mean(x), np.std(x), stats.skew(x), stats.kurtosis(x)])

    # ------------------- Relative entropy between distributions with n = 2, c = 1 as the reference distribution
        # NOTE: We use Ex-Gaussian Distributions here.
    agg_data["RE_n2c1_n2c1"] = np.array([RE_uni_exg_numerical(params_px = agg_data.exg_params_n2c1[i], params_pref = agg_data.exg_params_n2c1[i]) for i in range(1,agg_data.shape[0]+1)])
    agg_data["RE_n2c2_n2c1"] = np.array([RE_uni_exg_numerical(params_px = agg_data.exg_params_n2c2[i], params_pref = agg_data.exg_params_n2c1[i]) for i in range(1,agg_data.shape[0]+1)])
    agg_data["RE_n4c1_n2c1"] = np.array([RE_uni_exg_numerical(params_px = agg_data.exg_params_n4c1[i], params_pref = agg_data.exg_params_n2c1[i]) for i in range(1,agg_data.shape[0]+1)])
    agg_data["RE_n4c2_n2c1"] = np.array([RE_uni_exg_numerical(params_px = agg_data.exg_params_n4c2[i], params_pref = agg_data.exg_params_n2c1[i]) for i in range(1,agg_data.shape[0]+1)])

    # ------------------ Relative Entropy All by All matrix ------------------------------------------------------

    agg_data["RE_all"] = 0
    agg_data["RE_all"] = agg_data["RE_all"].astype(object)

    for i_player in range(1, agg_data.shape[0]+1):
       dist_params =  agg_data.loc[i_player, ['exg_params_n2c1', 'exg_params_n2c2', 'exg_params_n4c1', 'exg_params_n4c2']]

       # for an all by re matrix
       re_matrix = np.zeros([4, 4])
       
       for i_x in range(4): # p(x) index
         for i_ref in range(4): # pref index
          re_matrix[i_x, i_ref] = RE_uni_exg_numerical(params_px = dist_params[i_x], params_pref = dist_params[i_ref])
       agg_data.at[i_player, "RE_all"] = re_matrix

    # ------------------- Effect size through avg relative entropy between conditions ----

    # We estimate setsize by the relative entropy between distributions across set size, for each of congruency conditions.
    # Four terms (instead of two) as relative is asymmetric. 
    agg_data["avg_setsize_effect"] = agg_data["RE_all"].apply(lambda x: (x[2,0] + x[3,1] + x[0,2] + x[1,3])/4)

    # We estimate setsize by the relative entropy between distributions across congruency conditions, for each set size.
    agg_data["avg_simon_effect"] = agg_data["RE_all"].apply(lambda x: (x[1,0] + x[0,1] + x[2,3] + x[3,2])/4)

    # Set size effect to Simon effect ratio (need a better, self-explanatory name)
    # higher means relatively more affected by setsize than stimulus response incompatibility.
    agg_data["effect_ratio"] = agg_data["avg_setsize_effect"]/agg_data["avg_simon_effect"] 

    # agg_data.sort_values(by = ["effect_ratio"]).effect_ratio
    # [30,60], ...[The middle is possibly a mixture of strategy groups: 26 (no effect, surprisingly), 57 (both effect + unexplained). Future studies: What strategies would explain such changes]...., [29, 54]
    
    # ---------------------- Accuracy --------------------------

    agg_data["mean_acc_n2c1"] = agg_data.sorted_cor_n2c1.apply(lambda x: np.sum(x)/x.shape[0])
    agg_data["mean_acc_n2c2"] = agg_data.sorted_cor_n2c2.apply(lambda x: np.sum(x)/x.shape[0])
    agg_data["mean_acc_n4c1"] = agg_data.sorted_cor_n4c1.apply(lambda x: np.sum(x)/x.shape[0])
    agg_data["mean_acc_n4c2"] = agg_data.sorted_cor_n4c2.apply(lambda x: np.sum(x)/x.shape[0])

    # ----------- Experimental effects though differences of Ex-Gaussian parameters ----------

    # mu, simon effect (different S-R compatibility with same setsize)
    agg_data["d_mu_n2"] = (agg_data.exg_params_n2c2.apply(lambda x: x[1]) - agg_data.exg_params_n2c1.apply(lambda x: x[1]))/agg_data.exg_params_n2c1.apply(lambda x: x[1])*100
    agg_data["d_mu_n4"] = (agg_data.exg_params_n4c2.apply(lambda x: x[1]) - agg_data.exg_params_n4c1.apply(lambda x: x[1]))/agg_data.exg_params_n4c1.apply(lambda x: x[1])*100
    
    # mu, setsize effect (different setsizes with same SR compatibility)
    agg_data["d_mu_c1"] = (agg_data.exg_params_n4c1.apply(lambda x: x[1]) - agg_data.exg_params_n2c1.apply(lambda x: x[1]))/agg_data.exg_params_n2c1.apply(lambda x: x[1])*100
    agg_data["d_mu_c2"] = (agg_data.exg_params_n4c2.apply(lambda x: x[1]) - agg_data.exg_params_n2c2.apply(lambda x: x[1]))/agg_data.exg_params_n4c1.apply(lambda x: x[1])*100

    # sigma, simon effect
    agg_data["d_sigma_n2"] = (agg_data.exg_params_n2c2.apply(lambda x: x[2]) - agg_data.exg_params_n2c1.apply(lambda x: x[2]))/agg_data.exg_params_n2c1.apply(lambda x: x[2])*100
    agg_data["d_sigma_n4"] = (agg_data.exg_params_n4c2.apply(lambda x: x[2]) - agg_data.exg_params_n4c1.apply(lambda x: x[2]))/agg_data.exg_params_n4c1.apply(lambda x: x[2])*100
    # sigma, setsize effect
    agg_data["d_sigma_c1"] = (agg_data.exg_params_n4c1.apply(lambda x: x[2]) - agg_data.exg_params_n2c1.apply(lambda x: x[2]))/agg_data.exg_params_n2c1.apply(lambda x: x[2])*100
    agg_data["d_sigma_c2"] = (agg_data.exg_params_n4c2.apply(lambda x: x[2]) - agg_data.exg_params_n2c2.apply(lambda x: x[2]))/agg_data.exg_params_n4c1.apply(lambda x: x[2])*100

    # tau, simon effect
    agg_data["d_tau_n2"] = (agg_data.exg_params_n2c2.apply(lambda x: x[0]*x[2]) - agg_data.exg_params_n2c1.apply(lambda x: x[0]*x[2]))/agg_data.exg_params_n2c1.apply(lambda x: x[0]*x[2])*100
    agg_data["d_tau_n4"] = (agg_data.exg_params_n4c2.apply(lambda x: x[0]*x[2]) - agg_data.exg_params_n4c1.apply(lambda x: x[0]*x[2]))/agg_data.exg_params_n4c1.apply(lambda x: x[0]*x[2])*100

    # tau, setsize effect
    agg_data["d_tau_c1"] = (agg_data.exg_params_n4c1.apply(lambda x: x[0]*x[2]) - agg_data.exg_params_n2c1.apply(lambda x: x[0]*x[2]))/agg_data.exg_params_n2c1.apply(lambda x: x[0]*x[2])*100
    agg_data["d_tau_c2"] = (agg_data.exg_params_n4c2.apply(lambda x: x[0]*x[2]) - agg_data.exg_params_n2c2.apply(lambda x: x[0]*x[2]))/agg_data.exg_params_n4c1.apply(lambda x: x[0]*x[2])*100


    agg_data.to_pickle("agg_RT_data_exp2_concise_pickle.txt")

else:
    
    agg_data = pd.read_pickle("Datasets/OurStudy2_Mittelstadt_exp2/agg_RT_data_exp2_concise_pickle.txt")


print(agg_data.columns.shape)
print(agg_data.columns)

print(agg_data.head())
```

### Case 2: Experimental effect through RE between all pairs of conditions

Instead of using the average effects, we can use pairwise comparisons of distributions. Six unique pairs of conditions --> the experimental effect on each individual as a point vector of six relative entropy values. In this case, the partitions are determined based on the eigenvalues and eigenvectors in the six-dimensional space, compared to the 2D effect space used earlier.


```python
def create_flat_RE_array_from_matrix(given_matrix):
    flattened_array = np.array([])
    for row_i in range(1, len(given_matrix)):
        temp = given_matrix[row_i, :row_i]
        # print(temp)
        flattened_array = np.concatenate([flattened_array, temp])
    return(flattened_array)
# create_flat_RE_array_from_matrix(agg_data.RE_all.iloc[2])
```
